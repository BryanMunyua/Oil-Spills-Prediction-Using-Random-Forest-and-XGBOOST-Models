DATA CLEANING

Replace missing categorical variables with the most common value in the column

```python
# Load the data into a DataFrame
df = pd.read_csv('database.csv')

categorical_columns = ['Cause Category', 'Pipeline Type', 'Liquid Type', 'Liquid Subtype',
                       'Pipeline Location', 'Pipeline Shutdown',
                       'Liquid Ignition', 'Liquid Explosion', 'Accident State']

# Iterate through the categorical columns and replace NaN with the mode
for col in categorical_columns:
    if col in df.columns:  # Ensure the column exists in the DataFrame
        most_common_value = df[col].mode()[0]  # Get the mode (most frequent value)
        df[col].fillna(most_common_value, inplace=True)  # Replace NaN with the mode

# Confirm changes
print(df[categorical_columns].isnull().sum())  # Verify no missing values remain in these columns
```

Remove outliers

```python
def removeOutliers(df):
    # Select only numeric columns for outlier detection
    numeric_df = df.select_dtypes(include=['float64', 'int64'])

    # Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numeric column
    Q1 = numeric_df.quantile(0.25)
    Q3 = numeric_df.quantile(0.75)
    IQR = Q3 - Q1  # Interquartile range

    # Filter out rows with outliers in numeric columns only
    numeric_df_clean = numeric_df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]

    # Combine the non-numeric columns back with the cleaned numeric data
    df = pd.concat([numeric_df_clean, df.select_dtypes(exclude=['float64', 'int64'])], axis=1)

    # Drop any rows with NaN values introduced by the filtering
    df_clean = df.dropna()

    return df_clean

    # Display the resulting DataFrame
    print(df_clean)

df_clean = removeOutliers(df)
```

Parse the Date/Time

```python
# Date/Time Parsing
def dateTimeParsing(df):
    # Ensure the 'Accident Date/Time' column is in datetime format
    df['Accident Date/Time'] = pd.to_datetime(df['Accident Date/Time'], errors='coerce')

    # Extract the month from the date
    df['Accident Month'] = df['Accident Date/Time'].dt.month

    # Extract the day of the week (0=Monday, 6=Sunday)
    df['Accident Day of Week'] = df['Accident Date/Time'].dt.dayofweek

    # Define a function to map months to seasons
    def get_season(month):
        if month in [12, 1, 2]:
            return 'Winter'
        elif month in [3, 4, 5]:
            return 'Spring'
        elif month in [6, 7, 8]:
            return 'Summer'
        elif month in [9, 10, 11]:
            return 'Fall'
        return None

    # Apply the season function to create a new 'Season' column
    df['Accident Season'] = df['Accident Month'].apply(get_season)

    # Display the transformed DataFrame with new columns
    print(df[['Accident Date/Time', 'Accident Month', 'Accident Day of Week', 'Accident Season']].head())

    return df

df_clean = dateTimeParsing(df_clean)
```

Encode categorical variables

```python
def encodeCategoricalVariables(df):
    # List of categorical columns to encode
    categorical_columns = ['Cause Category', 'Pipeline Type', 'Liquid Type','Liquid Subtype',
                           'Pipeline Location', 'Accident Season', 'Pipeline Shutdown', 'Liquid Ignition',
                           'Liquid Explosion','Accident State']

    # Encode each categorical column with integer codes
    for column in categorical_columns:
        df[column], _ = pd.factorize(df[column])

    # Display the DataFrame with encoded columns
    print("DataFrame after Numerical Encoding:")
    print(df.head())

    return df

df_clean = encodeCategoricalVariables(df_clean)
```

```python
def combineColumns(df, dividedCol, divisorCol, newcol):
    # Safely divide dividedCol by divisorCol, setting result to 0 where divisorCol is zero
    df[newcol] = np.where(df[divisorCol] != 0, df[dividedCol] / df[divisorCol], 0)

    # Display the updated dataframe
    print(df)

    return df

combineColumns(df_clean, 'Pipeline Type', 'Liquid Type','pipelineCombined')
combineColumns(df_clean, 'Accident Latitude', 'Accident Longitude','latLongLocation')
combineColumns(df_clean, 'Liquid Ignition', 'Liquid Explosion','liquidCombined')
combineColumns(df_clean, 'Cause Category', 'Pipeline Type','causecategoryPipelineTypeCombined')
combineColumns(df_clean, 'Pipeline Type', 'Pipeline Shutdown','TypeAndShutdownCombined')
```

Chi Square Test

```python
def chisquareTest(df, col1, col2):
    # Define the categorical variables
    cause_category = df[col1]
    spill_occurrence = df[col2]

    # Create a contingency table
    contingency_table = pd.crosstab(cause_category, spill_occurrence)

    # Perform the Chi-Square Test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Print the test results
    print("Chi-Square Statistic:", chi2)
    print("p-value:", p)
    print("Degrees of Freedom:", dof)
    print("Expected Frequencies:\n", expected)

    # Interpretation
    if p < 0.05:
        print(f"There is a significant association between {col1} and Spill Occurrence.")
    else:
        print(f"There is no significant association between {col1} and Spill Occurrence.")

chisquareTest(df_clean,'Cause Category','Target')
chisquareTest(df_clean,'Pipeline Type','Target')
chisquareTest(df_clean,'Liquid Type','Target')
```

ANOVA Test

```python
def ANOVAtest(df, col1, col2):
    # Group data by 'Cause Category' and extract 'Property Damage Costs' for each category
    groups = [df[col1][df[col2] == category] for category in df[col2].unique()]

    # Perform one-way ANOVA
    f_statistic, p_value = f_oneway(*groups)

    # Print the ANOVA results
    print("F-Statistic:", f_statistic)
    print("p-value:", p_value)

    # Interpretation
    if p_value < 0.05:
        print("There is a significant difference in Property Damage Costs across different Cause Categories.")
    else:
        print("There is no significant difference in Property Damage Costs across different Cause Categories.")

ANOVAtest(df_clean, 'Property Damage Costs', 'Cause Category')
```

Random Forest Model with AUC and ROC

```python
def model():
    # Define the target variable and features
    target_column = 'Target'
    feature_columns = ['Cause Category', 'Pipeline Type', 'Liquid Type', 'Liquid Subtype',
                       'Pipeline Location', 'Accident Season', 'Pipeline Shutdown',
                       'Liquid Ignition', 'Liquid Explosion', 'Accident State', 'Accident Latitude',
                       'Accident Longitude', 'Accident Day of Week', 'causecategoryPipelineTypeCombined',
                       'latLongLocation', 'TypeAndShutdownCombined']

    # Separate features and target variable
    X = df_clean[feature_columns]
    y = df_clean[target_column]

    # Encoding categorical features
    for col in X.select_dtypes(include=['object']).columns:
        X[col] = LabelEncoder().fit_transform(X[col])

    # Handle missing values (filling with 0 here, but other strategies can be used)
    X = X.fillna(0)

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    # Initialize the model (using Random Forest here)
    model = RandomForestClassifier(n_estimators=1000, random_state=42)

    # 5-fold cross-validation
    cv_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]  # Get probabilities for the positive class

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_pred_proba)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print("Model Accuracy:", accuracy)
    print("Cross-Validation Accuracy Scores:", cv_scores)
    print("AUC Score:", auc_score)
    print("\nConfusion Matrix:\n", conf_matrix)
    print("\nClassification Report:\n", class_report)

    # Plot the ROC Curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    plt.figure()
    plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC Curve (AUC = {auc_score:.2f})")
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for random guessing
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

model()
```
